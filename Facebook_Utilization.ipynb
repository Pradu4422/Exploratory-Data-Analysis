Question 1. What is the range of age of people using facebook? Is it normally distributed?

```python
data['age'].min()
```

```python
data['age'].max()
```

```python
# Insert your code here...
data['age-group'] = pd.cut(data['age'], [0, 18, 25, 40, 60, 90, 113], labels = ['0-18', '18-25', '26-40', '41-60', '61-90', '91-113'], include_lowest=True)
```

```python
data['age-group'].value_counts().plot(kind='bar',figsize=(10,5), colormap='Dark2', fontsize=13,yticks=np.arange(0,30000,5000))
plt.title('Breakdown of age-groups across User count')
plt.xlabel('Age-Group')
plt.ylabel('User Count')
plt.savefig('age-group vs user_count.jpg')
```

```python
data['age'].plot(kind='kde', figsize=(10, 7), color='brown')
plt.title('Distribution of user count across Age-Group')
plt.savefig('Distribution of user count across Age-Group.jpg')
```

Answer 1:
- There are existing users has range between 13 to 113.
- Approx 26000 of users having 18-25 age.
- Approx 23500 of users having 26-40 age. It means teenagers has more impact on facebook utilization.
- Approx 18500 of users having 41 to 60 age. It means Adult has also more impact on facebook utilization but leass than 18-40.
- Approx 17000 of users has age below than 18. 
- After 61 years of age there are decresing number in users meaning oldage don't use social media much which is obvious.
- Age is not normally distributed in dataset. It has three peaks. Also all the data is right skewed.

Question 2. Which age group has more accounts on facebook with respect to Gender?

```python
data['age-group'].value_counts().plot(kind='pie', fontsize=14, autopct='%.2f%%', figsize=(10,10),wedgeprops=dict(width=0.20), legend=True, startangle=160, shadow=True , cmap='plasma' )

plt.title('User counts % across different Age-Groups\n',fontsize=30)
plt.ylabel('Age-Group\n')
plt.legend(loc="center")
plt.savefig("User counts % across different age-groups.jpg")             # The image will be saved in the directory of this notebook.

```

```python
agegroup_gender = pd.pivot_table(data, index='age-group',values = 'friend_count', columns='gender', aggfunc='count')
agegroup_gender
```

```python
agegroup_gender[['female','male']].plot(kind='bar', stacked=True, figsize=(15,8))
plt.title('User counts across Age-Groups & Gender\n',fontsize=30)
plt.xlabel('Age-Groups')
plt.ylabel('User counts\n')
plt.xticks(rotation=0)
plt.savefig("User counts across Age-Groups & Gender.jpg")  
```

Question 3: Which age-group is more active on Facebook?

```python
colors = ['c', 'b','r','y','g']
data.groupby(['age-group'])['friend_count', 'friendships_initiated'].sum().plot(kind='pie', colors= colors, pctdistance=0.85, fontsize=14, subplots=True, autopct='%.2f%%', figsize=(30,30),wedgeprops=dict(width=0.40), legend=False, startangle=160, shadow=True)
plt.title('Active user counts % across different Age-Groups\n',fontsize=30)
plt.legend(loc="best")
plt.savefig("Active age-groups on facebook.jpg")             # The image will be saved in the directory of this notebook.

plt.show()
```

Answer 3: 
- 0-18 and 18-25 Age-group is more active on facebook. 
- 26-40 Age-group has also good contribution in facebook.
- Each age-group has highly correlation between friend_Count, friendships_initiated are highly correlated.
